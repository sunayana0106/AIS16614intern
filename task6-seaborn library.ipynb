# Python Seaborn library    
* Visualize Distributions
* Seaborn is a library that uses Matplotlib underneath to plot graphs.
* It will be used to visualize random distributions.
* Seaborn is a powerful Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics.
* It is particularly well-suited for visualizing complex datasets and making complex visualizations simple to create.s.

* Distplots    
      Distplot stands for distribution plot, it takes as input an array and plots a curve corresponding to the distribution of points in the array.

```python
# Plotting a Distplot
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot([0, 1, 2, 3, 4, 5])
plt.show()
```

```python
# Plotting a Distplot Without the Histogram
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot([0, 1, 2, 3, 4, 5], hist=False)
plt.show()
```

* Scatter plot with regression line

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")

# Create scatter plot with regression line
sns.lmplot(x="total_bill", y="tip", data=tips, aspect=1.5)

# Customize plot
plt.title('Scatter Plot with Regression Line: Total Bill vs Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.show()
```

* Box Plot with Subplots

```python
# Load dataset
tips = sns.load_dataset("tips")

# Create subplots
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Box plot by day
sns.boxplot(x="day", y="total_bill", data=tips, ax=ax[0])
ax[0].set_title('Total Bill Distribution by Day')

# Box plot by time
sns.boxplot(x="time", y="total_bill", data=tips, ax=ax[1])
ax[1].set_title('Total Bill Distribution by Time')

plt.show()
```

* FacetGrid with Histogram

```python
# Load dataset
tips = sns.load_dataset("tips")

# Create a FacetGrid
g = sns.FacetGrid(tips, col="sex", row="time", margin_titles=True)
g.map(sns.histplot, "total_bill", bins=20)

# Show plot
plt.show()
```

*  Line Plot with Multiple Lines

```python
# Load dataset
fmri = sns.load_dataset("fmri")

# Create line plot
sns.lineplot(x="timepoint", y="signal", hue="region", style="event", data=fmri, markers=True, dashes=False)

# Customize plot
plt.title('FMRI Signal over Time')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.show()
```

* Histogram

```python
# Load dataset
tips = sns.load_dataset("tips")

# Create histogram
sns.histplot(tips["total_bill"], bins=20, kde=True)

# Customize plot
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')
plt.show()
```

* Bar Plot 

```python
# Load dataset
titanic = sns.load_dataset("titanic")

# Create bar plot
sns.barplot(x="class", y="survived", hue="sex", data=titanic)

# Customize plot
plt.title('Survival Rate by Class and Gender')
plt.xlabel('Class')
plt.ylabel('Survival Rate')
plt.show()
```

* Residual Plot

```python
# Load dataset
tips = sns.load_dataset("tips")

# Create residual plot
sns.residplot(x="total_bill", y="tip", data=tips, lowess=True, color="g")

# Customize plot
plt.title('Residual Plot of Total Bill and Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Residuals')
plt.show()
```

* Time Series Plot

```python
# Load dataset
flights = sns.load_dataset("flights")

# Create line plot for time series data
sns.lineplot(x="year", y="passengers", data=flights, ci=None, marker='o')

# Customize plot
plt.title('Number of Passengers Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.show()
```
